## Вычисляемые поля

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Бестселлеры.csv")
data
```

```python
data.columns = ["name", "author", "user_rating", "reviews", "price", "year", "genre"]
```

Самый простой способ расчёта вычисляемого поля

```python
data["sale"] = data["price"] * 0.8
```

Функция apply

```python
def get_sale(row):
    if row["genre"] == "Non Fiction" and row["year"] < 2015:
        return row["price"] * 0.5
    if row["user_rating"] < 4.3:
        return row["price"] * 0.8
    return row["price"]
```

```python
data["sale"] = data.apply(get_sale, axis=1)
data
```

Функция map

```python
data["name"].map(len)
```

```python
def my_func(item):
    if len(item) > 20:
        return "Длинное имя книги"
    else:
        return "Короткое имя книги"
```

```python
data["name"].map(my_func)
```

```python
data["name_len"] = data["name"].map(len)
```
