## Операции над численными столбцами

```python
import pandas as pd
```

```python
data = pd.DataFrame(
    {
        "a": [3, 1, 2, 3, 4],
        "b": [1, 6, 4, 6, 9],
    }
)
data
```

Операции агрегации

```python
data["a"].mean()  # среднее значение
```

```python
data["a"].sum()  # Сумма
```

```python
data["a"].min()  # Минимальное значение
```

```python
data["a"].max()  # Максимальное значение
```

Трансформирующие операторы

```python
data["a"] + 5
```

```python
data["a"] + data["b"]  # поэлементное сложение столбцов
```

```python
data["result"] = data["a"].diff()  # Из последующего вычитается предыдущее
data
```

```python
data["result"] = data["a"].cumsum()  # Где на каждом шаге расчитывается сумма предыдущих элементов
data
```
