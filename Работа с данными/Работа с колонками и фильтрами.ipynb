Но вторым способом мы не сможем обратиться к столбцу User Rating. Поэтому столбцы принято переименовывать, чтобы в них небыло пробелов.

```python
data = data.rename({"User Rating": "user_rating"}, axis=1)
```

```python
data.columns = ["name", "author", "user_rating", "reviews", "price", "year", "genre"]
```

Фильтрация по логическим условиям

```python
data[data["user_rating"] > 4.7]
```

```python
data[(data["user_rating"] > 4.7) & (data["year"] == 2019) | (data["genre"] == "Non Fiction") & ~(data["author"] == "Michelle Obama")]
```

```python
data.query("user_rating > 4.7 and year == 2019 or genre == 'Non Fiction' and not author == 'Michelle Obama'")
```

## Изменение значений

Если мы хотим сделать скидку 20% для книг с рейтингом меньше 4, то данный пример не позволит изменить значения

```python
data[data["user_rating"] < 4]['price'] = data[data["user_rating"] < 4]['price'] * 0.8
```

Нужно использовать команду loc

```python
data.loc[data["user_rating"] < 4, "price"] = data[data["user_rating"] < 4]["price"] * 0.8
```

Очень важно, чтобы индексы у левой и правой стороны совпадали. Код ниже работать правильно не будет

```python
data.loc[data["user_rating"] < 4, "price"] = data[data["user_rating"] < 4]["price"].reset_index()
```
